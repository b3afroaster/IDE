```python
def concat_user_files(path):
    user_lst = os.listdir(path).sort()
    name1 = user_lst[0]
    user_lst = user_lst.pop(0)
    fullname = path + '/' + name1
    user_files = pd.read_csv(fullname)
    for elements in user_lst:
        fullpath = path + '/' + element
        file0 = pd.read_csv(fullpath)
        user_files = pd.concat([user_files, file0])
    result = user_files.drop_duplicates(ignore_index=True)
    return result
```
