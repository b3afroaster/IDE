```python
recepies = pd.read_json('https://lms-cdn.skillfactory.ru/assets/courseware/v1/92fd198fd3eccc09a8c3498e9dd25588/asset-v1:Skillfactory+DSMED+2023+type@asset+block/recipes.json')
```

```python
recepies.head()
```

```python
ingr = recepies['ingredients']
def func(stuff):
    jst_set = set()
    for list in stuff:
        for i in list:
            jst_set.add(i)
    return jst_set
```

```python
all_ingredients  = func(ingr)
len(all_ingredients)
```

```python
def contains(ingredient_list): 
    # Если ингредиент есть в текущем блюде,
    if ingredient_name in ingredient_list:   
        # возвращаем значение 1
        return 1 
    # Если ингредиента нет в текущем блюде,
    else: 
        # возвращаем значение 0
        return 0
```

```python
for ingredient_name in all_ingredients: 
    # В DataFrame cоздаем столбец с именем текущего ингредиента 
    # и заполняем его единицами и нулями,
    # используя ранее созданную функцию contains
    recepies[ingredient_name] = recepies['ingredients'].apply(contains)
```

```python
display(recepies)
```

```python
# Заменяем список ингредиентов в рецепте на их количество 
recepies['ingredients'] = recepies['ingredients'].apply(len) 
```

```python
id_list = list(recepies['id'])
```

```python

ingredients=list(recepies.columns)[3:]
```

```python
new_recepies = []
```

```python
def make_list(dframe_id):
    made_list = []
    for ingr in ingredients:
        if dframe_id[ingr].item()==1:
            made_list.append(ingr) 
    return made_list
```

```python
def make_list(row): 
    # Создаем пустой список ингредиентов текущего блюда
    ingredient_list=[] 
    # Последовательно перебираем ингредиенты из реестра
    for ingredient in ingredients: 
        # Если текущий ингредиент входит в состав текущего блюда
        if row[ingredient].item()==1: 
            # Добавляем ингредиент в список ингредиентов текущего блюда
            ingredient_list.append(ingredient) 
    # Возвращаем сформированный список ингредиентов
    return ingredient_list
```

```python
for uni_id in id_list:
    dish_dict = {}
    #определяю ключ id в словаре
    dish_dict['id'] = uni_id
    dish_frame = recepies[recepies['id'] == uni_id]
    #определяю ключ cuisine в словаре
    dish_dict['cuisine'] = dish_frame['cuisine']
    #определяю ключ ingredients
    ingr_list = make_list(dish_frame)
    dish_dict['ingredients'] = ingr_list
    #вношу словарь для блюда в список
    new_recepies.append(dish_dict)
    
```

```python
# Импорт модуля json
import json 
# Функция dumps() модуля json сериализирует объект Python в строку формата JSON. 
new_recipes = json.dumps(new_recipies) 

# Откроем файл new_recipes.json для записи
with open("data/new_recipies.json", "w") as write_file: 
    # Записываем содержимое подготовленные данные в файл
    write_file.write(new_recipies)
```
